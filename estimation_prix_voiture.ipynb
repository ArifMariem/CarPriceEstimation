### Randomforest

```python
n_estimators = [5,20,50,100] # number of trees in the random forest
max_features = ['auto', 'sqrt'] # number of features in consideration at every split
max_depth = [int(x) for x in np.linspace(10, 120, num = 12)] # maximum number of levels allowed in each decision tree
min_samples_split = [2, 6, 10] # minimum sample number to split a node
min_samples_leaf = [1, 3, 4] # minimum sample number that can be stored in a leaf node
bootstrap = [True, False] # method used to sample data points

random_grid = {'n_estimators': n_estimators,

'max_features': max_features,

'max_depth': max_depth,

'min_samples_split': min_samples_split,

'min_samples_leaf': min_samples_leaf,

'bootstrap': bootstrap}
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
from sklearn.model_selection import RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator = rf,param_distributions = random_grid,
               n_iter = 100, cv = 5, verbose=2, random_state=35, n_jobs = -1)
rf_random.fit(x_train, y_train)
```

```python
rf_random.fit(x_train, y_train)
print ('Random grid: ', random_grid, '\n')
# print the best parameters
print ('Best Parameters: ', rf_random.best_params_, ' \n')
```

```python
randmf = RandomForestRegressor(n_estimators = 50, min_samples_split = 6, min_samples_leaf= 1, max_features = 'sqrt', max_depth= 40, bootstrap=False , criterion="mse") 
randmf.fit( x_train, y_train) 
```

```python
from sklearn.metrics import r2_score 
from sklearn import metrics


y_pred_rf1 = pd.DataFrame( { "actual": y_test, 
"predicted_prob": randmf.predict( 
( x_test ) ) } ) 
#y_pred_rf1
y_predd = randmf.predict(x_test)
x0=r2_score(y_test, y_predd)
x1=metrics.mean_absolute_error(y_test,y_predd)
x2 =metrics.mean_squared_error(y_test, y_predd)
x3=np.sqrt(metrics.mean_squared_error(y_test, y_predd))
#print('R^2: ', adj(y_test, y_pred))
print('R^2: ', r2_score(y_test, y_predd))
print('MAE:', metrics.mean_absolute_error(y_test,y_predd))
print('MSE:', metrics.mean_squared_error(y_test, y_predd))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_predd)))
```
